# Q1

```python
def is_symmetrical(n):
    s=str(n)
    if(s==s[::-1]):
        return True
    else:
        return False
```

```python
is_symmetrical(7227)
```

```python
is_symmetrical(12567)
```

```python
is_symmetrical(44444444)
```

```python
is_symmetrical(9939)
```

```python
is_symmetrical(1112111)
```

# Q2

```python
def multiply_nums(string):
    mul=1
    l=string.split(', ')
    for i in l:
        mul=mul*int(i)
    return mul    
```

```python
multiply_nums('2, 3')
```

```python
multiply_nums('1, 2, 3, 4')
```

```python
multiply_nums('54, 75, 453, 0')
```

```python
multiply_nums('10, -2')
```

# Q3

```python
def square_digits(n):
    s=str(n)
    a=''
    for i in s:
        a=a+str(int(i)**2)
    return int(a)    
```

```python
square_digits(9119)
```

```python
square_digits(2483)
```

```python
square_digits(3212)
```

# Q4

```python
def setify(l):
    s=set(l)
    l=list(s)
    l.sort()
    return l
```

```python
l=[1,3,3,5,5]
setify(l)
```

```python
l=[4,4,4,4]
setify(l)
```

```python
l=[5,7,8,9,10,15]
setify(l)
```

```python
l=[3,3,3,2,1]
setify(l)
```

# Q5

```python
def mean(n):
    s=str(n)
    sum=0
    l=len(s)
    for i in s:
        sum=sum+int(i)
    return int(sum/l)    
```

```python
mean(42)
```

```python
mean(12345)
```

```python
mean(666)
```
