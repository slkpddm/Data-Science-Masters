# Q1

```python
'''
Question1
Create a function that takes three integer arguments (a, b, c) and returns the amount of
integers which are of equal value.
Examples
equal(3, 4, 3) ➞ 2
equal(1, 1, 1) ➞ 3
equal(3, 4, 1) ➞ 0
Notes
Your function must return 0, 2 or 3.
'''
print()
```

```python
def equal(x, y, z):
    l=[]
    c=[]
    l.append(x)
    l.append(y)
    l.append(z)
    c.append(l.count(l[0]))
    c.append(l.count(l[1]))
    if(max(c)>1):
        return max(c)
    else:
        return 0       
```

```python
equal(3,4,3)
```

```python
equal(1, 1, 1)
```

```python
equal(3, 4, 1)
```

# Q2

```python
def dict_list(dict1):
    l=[]
    for i in dict1.keys():
        l.append(i)
    l.sort()
    dict2={}
    for i in l:
        dict2[i]=dict1[i]
    return list(dict2.items())    
```

```python
d={'D':1,'B':2,'C':3}
dict_list(d)
```

```python
d1={'likes':2,'dislikes':3,'followers':10}
dict_list(d1)
```

# Q3

```python
def map_dict(l):
    dict1={}
    for i in l:
        dict1[i]=i.upper()
    return dict1    
    
```

```python
l=['p','s']
map_dict(l)
```

```python
l=['a','b','c']
map_dict(l)
```

```python
l=['a','v','y','z']
map_dict(l)
```

# Q4

```python
def vow_replace(string,char):
    v=['a','e','i','o','u']
    s=''
    for i in string:
        if(i in v):
            s=s+char
        else:
            s=s+i
    return s      
```

```python
vow_replace('apples and bananas','u')
```

```python
vow_replace('cheese casserole','o')
```

```python
vow_replace('stuffed jalapeno poppers','e')
```

```python
def ascii_capitalize(string):
    s=''
    for i in string:
        if(ord(i)%2==0):
            s=s+i.upper()
        else:
            s=s+i.lower()
    return s        
```

```python
ascii_capitalize('to be or not to be!')
```

```python
ascii_capitalize('THE LITTLE MERMAID')
```

```python
ascii_capitalize('Oh what a beautiful morning.')
```
