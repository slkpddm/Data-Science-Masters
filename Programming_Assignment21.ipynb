# Q1

```python
def next_in_line(l,n):
    if(len(l)>0):
        l.pop(0)
        l.append(n)
        return l
    else:
        return 'No list has been selected'
```

```python
next_in_line([5,6,7,8,9],1)
```

```python
next_in_line([7,6,3,23,17],10)
```

```python
next_in_line([1,10,20,42],6)
```

```python
next_in_line([],6)
```

# Q2

```python
def get_budgets(*dicts):
    bgt=0
    for i in dicts:
        bgt=bgt+i['budget']
    return bgt    
        
```

```python
d1={'name':'John','age':21,'budget':23000}
d2={'name':'Steve','age':32,'budget':40000}
d3={'name':'Martin','age':16,'budget':2700}
get_budgets(d1,d2,d3)
```

```python
d1={'name':'John','age':21,'budget':29000}
d2={'name':'Steve','age':32,'budget':32000}
d3={'name':'Martin','age':16,'budget':1600}
get_budgets(d1,d2,d3)
```

# Q3

```python
def alphabet_soup(string):
    l=list(string)
    l.sort()
    s=''
    for i in l:
        s=s+i
    return s    
```

```python
alphabet_soup('hello')
```

```python
alphabet_soup('edabit')
```

```python
alphabet_soup('hacker')
```

```python
alphabet_soup('geek')
```

```python
alphabet_soup('javascript')
```

# Q4

```python
def compound_interest(p,t,r,n):
    A = p*(1+(r/n))**(n*t)
    return round(A,2)
```

```python
compound_interest(100,1,0.05,1)
```

```python
compound_interest(3500,15,0.1,4)
```

```python
compound_interest(100000,20,0.15,365)
```

# Q5

```python
def return_only_integer(l):
    l1=[]
    for i in l:
        if(type(i)==int):
            l1.append(i)
    return l1        
```

```python
return_only_integer([9,2,'space','car','lion',16])
```

```python
return_only_integer(['hello',81,'basketball',123,'fox'])
```

```python
return_only_integer([10,'121',56,20,'car',3,'lion'])
```

```python
return_only_integer(['string',True,3.3,1])
```
