# Q1

```python
def double_char(string):
    s=''
    for i in string:
        s=s+2*i
    return s    
```

```python
double_char('String')
```

```python
double_char('Hello World!')
```

```python
double_char('1234!_ ')
```

# Q2

```python
def reverse(boolean):
    if(type(boolean)!=bool):
        return 'boolean expected'
    elif(boolean==True):
        return False
    else:
        return True
```

```python
reverse(True)
```

```python
reverse(False)
```

```python
reverse(0)
```

```python
reverse(None)
```

# Q3

```python
def num_layers(n):
    thick=0.0005
    for i in range(n):
        thick=thick*2
        
    return '{}m'.format(thick)   
```

```python
num_layers(1)
```

```python
num_layers(4)
```

```python
num_layers(21)
```

# Q4

```python
def index_of_caps(string):
    l=[]
    for i in string:
        if(i.isupper()):
            l.append(string.index(i))
    return l        
```

```python
index_of_caps('eDaBiT')
```

```python
index_of_caps('eQuINoX')
```

```python
index_of_caps('determine')
```

```python
index_of_caps('STRIKE')
```

```python
index_of_caps('sUn')
```

# Q5

```python
def find_even_nums(n):
    l=[i for i in range(1,n+1) if(i%2==0)]
    return l
```

```python
find_even_nums(8)
```

```python
find_even_nums(4)
```

```python
find_even_nums(2)
```
