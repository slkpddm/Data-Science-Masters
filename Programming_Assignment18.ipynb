# Q1

```python
def filter_list(l):
    for i in l:
        if(type(i)==int and i<0):
            return 'Enter the list with only non negative integers and strings'
            break
    else:
        lst=[]
        for i in l:
            if(type(i)==int):
                lst.append(i)
        return lst        
            
```

```python
filter_list([1,2,'a','b'])
```

```python
filter_list([1,'a','b',0,15])
```

```python
filter_list([1,2,'aasf','1','123',123])
```

```python
filter_list([-1,0,'skp',1,2,4,'dhoni'])
```

# Q2

```python
def reverse(string):
    s=''
    for i in string[::-1]:
        if(i.isupper()):
            s=s+i.lower()
        else:
            s=s+i.upper()
    return s        
    
```

```python
reverse('Hello World')
```

```python
reverse('ReVeRsE')
```

```python
reverse('Radar')
```

# Q3

```python
def unpack(l):
    first=l[0]
    middle=l[1:-1]
    last=l[-1]
    return 'first:{} middle:{} , last:{}'.format(first,middle,last)
```

```python
unpack([1,2,3,4,5,6])
```

# Q4

```python
def factorial(n):
    
    if(n==1 or n==0):
        return 1
    else:
        return (n*factorial(n-1))
```

```python
factorial(5)
```

```python
factorial(3)
```

```python
factorial(1)
```

```python
factorial(0)
```

# Q5

```python
def move_to_end(l,n):
    lst=[]
    l1=[]
    for i in l:
        if(i!=n):
            lst.append(i)
        else:
            l1.append(i)
    for i in l1:
        lst.append(i)
    return lst     
            
    
```

```python
move_to_end([1,3,2,4,4,1],1)
```

```python
move_to_end([7,8,9,1,2,3,4],9)
```

```python
move_to_end(['a','a','a','b'],'a')
```
