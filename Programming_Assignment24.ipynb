```python
def amplify(n):
    l=[]
    for i in range(1,n+1):
        if(i%4==0):
            l.append(i*10)
        else:
            l.append(i)
    return l        
```

```python
amplify(4)
```

```python
amplify(3)
```

```python
amplify(25)
```

```python
def unique(l):
    c=[]
    s=set(l)
    for i in s:
        if(l.count(i)==1):
            return i
```

```python
import math
class Circle:
    def __init__(self,radius):
        self.radius=radius
    def getArea(self):
        return math.pi*(self.radius**2)
    def getPerimeter(self):
        return 2*math.pi*self.radius    
```

```python
circy=Circle(11)
```

```python
circy.getArea()
```

```python
circy=Circle(4.44)
```

```python
circy.getPerimeter()
```

```python
def sort_by_length(l):
    le=[]
    ls=[]
    for i in l:
        le.append(len(i))
    le.sort()
    for i in le:
        for j in l:
            if(i==len(j)):
                ls.append(j)
    return ls            
```

```python
l=['Google','Apple','Microsoft']
sort_by_length(l)
```

```python
l=['Leonardo','Michelangelo','Raphael','Donatello']
sort_by_length(l)
```

```python
l=['Turing','Einstein','Jung']
sort_by_length(l)
```

```python
def is_triplet(a,b,c):
    l=[]
    l.append(a)
    l.append(b)
    l.append(c)
    l.sort()
    sum=l[0]**2+l[1]**2
    if(sum==l[2]**2):
        return True
    else:
        return False        
```

```python
is_triplet(3,4,5)
```

```python
is_triplet(13,5,12)
```

```python
is_triplet(1,2,3)
```
