# Q1

```python
def filter_list(l):
    lst=[]
    for i in l:
        if(type(i)==int):
            lst.append(i)
    return lst         
```

```python
l=[1,2,3,'a','b','c',4]
filter_list(l)
```

```python
l=['A',0,'Edabit',1729,'Python','1729']
filter_list(l)
```

```python
l=['Nothing','here']
filter_list(l)
```

# Q2

```python
def add_indexes(l):
    lst=[]
    for i in range(len(l)):
        lst.append(i+l[i])
    return lst   
```

```python
l=[0,0,0,0,0]
add_indexes(l)
```

```python
l=[1,2,3,4,5]
add_indexes(l)
```

```python
l=[5,4,3,2,1]
add_indexes(l)
```

# Q3

```python
import math
def cone_volume(h,r):
    v=math.pi*(r**2)*(h/3)
    return round(v,2)
```

```python
cone_volume(15,6)
```

```python
cone_volume(18,0)
```

```python
cone_volume(3,2)
```

# Q4

```python
def traingle(n):
    sum=0
    for i in range(0,n):
        sum=sum+(i+1)
    return sum    
```

```python
traingle(1)
```

```python
traingle(6)
```

```python
traingle(215)
```

```python
traingle(5)
```

```python
traingle(4)
```

```python
def missing_num(l):
    for i in range(1,11):
        if(i not in l):
            return i       
```

```python
missing_num([1,2,3,4,6,7,8,9,10])
```

```python
missing_num([7,2,3,6,5,9,1,4,8])
```

```python
missing_num([10,5,1,2,4,6,8,3,9])
```
