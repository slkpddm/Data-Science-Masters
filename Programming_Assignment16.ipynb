# Q1

```python
def shutter(string):
    s=''
    s=s+string[:2]+'... '+string[:2]+'... '+string+'?'
    return s
```

```python
shutter('incredible')
```

```python
shutter('enthusiastic')
```

```python
shutter('outstanding')
```

# Q2

```python
import math
from math import *
def radians_to_degrees(radians):
    deg = (180/pi)*radians
    return round(deg,1)
```

```python
radians_to_degrees(1)
```

```python
radians_to_degrees(20)
```

```python
radians_to_degrees(50)
```

# Q3

```python
def is_curzon(n):
    sum=2**n+1
    sum1=2*n+1
    if(sum%sum1==0):
        return True
    else:
        return False
```

```python
is_curzon(5)
```

```python
is_curzon(10)
```

```python
is_curzon(14)
```

# Q4

```python
import math
from math import *
def area_of_hexagon(s):
    A=1.5*sqrt(3)*pow(s,2)
    return round(A,1)
```

```python
area_of_hexagon(1)
```

```python
area_of_hexagon(2)
```

```python
area_of_hexagon(3)
```

# Q5

```python
def binary(n):
    s=''
    while(n>0):
        s=s+str(n%2)
        n=int(n/2)
    return s[::-1]    
```

```python
binary(1)
```

```python
binary(5)
```

```python
binary(10)
```
