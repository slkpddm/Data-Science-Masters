# Q1

```python
def division(n):
    for i in range(n):
        if(i%5==0 and i%7==0):
            yield i            
```

```python
division(100)
```

```python
list(division(100))
```

```python
for i in division(100):
    print(i,end='')   
    if(i!=list(division(100))[-1]):
        print(',',end='')
```

```python
def division(n):
    l=[]
    for i in range(n):
        if(i%5==0 and i%7==0):
            l.append(i)
    return l        
```

```python
for i in division(100):
    print(i,end='')
    if(i!=division(100)[-1]):
        print(',',end='')
```

# Q2

```python
def even(n):
    for i in range(n+1):
        if(i%2==0):
            yield i
```

```python
even(10)
```

```python
list(even(10))
```

```python
for i in even(10):
    print(i,end='')
    if(i!=list(even(10))[-1]):
        print(',',end='')
```

# Q3

```python
def f(n):
    if(n==0):
        return 0
    elif(n==1 or n==2):
        return 1
    else:
        return f(n-1)+f(n-2)
n=int(input())
for i in range(n+1):
    print(f(i),end='')
    if(i!=n):
        print(',',end='')    
```
